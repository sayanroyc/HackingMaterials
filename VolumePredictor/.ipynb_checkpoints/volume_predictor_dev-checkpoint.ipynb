## Test bench for tweaking outputs

```python
import os
import pandas as pd
from pymatgen import Structure
from matminer.learners.volume_predictor import VolumePredictor
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
import plotly.plotly as py

# use os.path.join to ensure proper '/' between folders and files on a different os
INITIAL_STRUCTURES_PATH = os.path.join(os.getcwd(), 'matminerStructures','initial')
```

### Import their matminer volume predictor into dataFrame

```python
cols_to_read = ['substituted', 'parent','vp Vf']
df=pd.read_csv('results_double_20170104.csv', usecols=cols_to_read)
```

### For each initial structure, run the volume predictor

```python
structures = []
initial_volumes = []
for f in os.listdir(INITIAL_STRUCTURES_PATH):
    if f.endswith(".cif"):
        fname = f.split('_')
        s = Structure.from_file(os.path.join(INITIAL_STRUCTURES_PATH, f))
        structures.append(s)
#         pred_vol_per_atom = VolumePredictor().predict(s)/s.num_sites
        pred_vol_per_atom = VolumePredictor().predict(s)
        initial_volumes.append([fname[0], fname[1], pred_vol_per_atom])
        print fname[0], fname[1], pred_vol_per_atom, s.volume 
        # we should find that pred_vol_per_atom == s.volume 
```

```python
i_v = pd.DataFrame(initial_volumes, columns=['substituted', 'parent', 'calc vp Vi'])
```

```python
out = i_v.merge(df, on=['substituted', 'parent'], left_index=True, right_index=True)
out['% diff'] = abs(out['calc vp Vi']-out['vp Vf'])/out['vp Vf']*100
out.sort_values('% diff', ascending=False)
```
